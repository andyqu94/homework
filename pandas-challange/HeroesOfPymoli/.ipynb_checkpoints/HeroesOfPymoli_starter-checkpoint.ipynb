```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data['SN'].unique())
display = pd.DataFrame({"Total Players":[total_players]})
display
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = len(purchase_data['Item ID'].unique())
average_price = purchase_data['Price'].mean()
number_purchase = len(purchase_data['Purchase ID'])
total_revenue = purchase_data['Price'].sum()
summary = pd.DataFrame({"Number of unique Items":[unique_item],"Average Price":[average_price],
                       "Number of Purchases":[number_purchase],"Total Revenue":[total_revenue]})
summary['Average Price'] = summary['Average Price'].map("${:.2f}".format)
summary['Total Revenue'] = summary['Total Revenue'].map("${:.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_table = purchase_data.drop_duplicates(subset='SN', keep = 'first')
Male = unique_table.loc[purchase_data['Gender'] == 'Male', : ]
Female = unique_table.loc[purchase_data['Gender'] == 'Female', : ]
total_male = len(Male)
total_female = len(Female)
total_length = len(unique_table)
other = total_length - total_male - total_female
male_percentage = (total_male/total_length)*100
female_percentage = (total_female/total_length)*100
other_percentage = (other/total_length)*100
percentage_summary = pd.DataFrame({"":['Male', 'Female', 'Other/Non-disclosed'],"total count":[total_male, total_female, other],
                                    "percentage of players":[male_percentage, female_percentage, other_percentage]})
percentage_summary['percentage of players'] = percentage_summary['percentage of players'].map('{:.2f}%'.format)
percentage_summary                                   
```


## Purchasing Analysis (Gender)

```python
group_gender = purchase_data.groupby(["Gender"])
purchase_count = group_gender["SN"].count()
purchase_Price = group_gender["Price"].mean()
purchase_Value = group_gender["Price"].sum()
unique_gender = unique_table.groupby(["Gender"])
avg_price = unique_gender["Price"].sum() / unique_gender["SN"].count()
purchasing_analysis = pd.DataFrame ({"Purchase Count":purchase_count,"Average Purchase Price":purchase_Price,
                                     "Total Purchase Value":purchase_Value, "Avg Total Purchase per Person":avg_price})
purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis['Total Purchase Value'] = purchasing_analysis["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis['Avg Total Purchase per Person'] = purchasing_analysis['Avg Total Purchase per Person'].map('${:2f}'.format)                                 
purchasing_analysis                                                                                                                
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,10,15,20,25,30,35,40,1000]
group =['<10','10 to 14','15 to 19','20 to 24','25 to 29','30 to 34','35 to 39','over 40']
replicate = purchase_data.copy()
replicate["Age Group"] = pd.cut(purchase_data["Age"], bins, labels= group)
group_age = replicate.groupby(["Age Group"])
unique_age = group_age['SN'].count()
total_age = purchase_data['SN'].count()
percentage_age = (unique_age/ total_age)*100
summary_age = pd.DataFrame({"Total Count":unique_age,"Percentage of Players":percentage_age})
summary_age['Percentage of Players'] = summary_age['Percentage of Players'].map('{:.2f}%'.format)
summary_age
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_count = group_age["SN"].count()
age_Price = group_age["Price"].mean()
age_Value = group_age["Price"].sum()
unique_age = unique_table.copy()
unique_age['Age Group'] = pd.cut(unique_age["Age"], bins, labels = group)
unique_replicate = unique_age.groupby('Age Group')
unique_df = unique_replicate["SN"].count() 
price_df = pd.merge(age_Value, unique_df, on="Age Group")
price_df["total average Price"] = price_df["Price"] / price_df["SN"]
first_df = pd.merge(age_count, price_df, on="Age Group")
second_df = pd.merge(first_df, age_Price, on="Age Group")
age_demographics = second_df.drop(['SN_y'],axis=1)
age_demographics = age_demographics.rename(columns={'SN_x':'Purchase Count','Price_x':'Average Purchase Price','Price_y':'Avg Total Perchase Per Person'})
age_demographics['Average Purchase Price'] = age_demographics['Average Purchase Price'].map('${:.2f}'.format)
age_demographics['total average Price'] = age_demographics['total average Price'].map('${:.2f}'.format)
age_demographics['Avg Total Perchase Per Person'] = age_demographics['Avg Total Perchase Per Person'].map('${:.2f}'.format)
age_demographics
```

```python
group_SN = purchase_data.groupby('SN')
purchase_count = group_SN['Purchase ID'].count()
purchase_total = group_SN['Price'].sum()
purchase_avg = purchase_total/purchase_count
top_spender = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":purchase_avg,"Total Purchase Value":purchase_total})
top_spender = top_spender.sort_values("Total Purchase Value",ascending = False)
top_spender['Average Purchase Price'] = top_spender['Average Purchase Price'].map('${:.2f}'.format)
top_spender['Total Purchase Value'] = top_spender['Total Purchase Value'].map('${:.2f}'.format)
top_spender.head()
```

## Top Spenders

```python
group = purchase_data.groupby(['Item ID','Item Name'])
group_count = group['SN'].count()
group_total = group['Price'].sum()
group_avg = (group_total / group_count)
group_price = group_avg*group_count
group_summary = pd.DataFrame({'purchase count':group_count,'item price':group_price,'total purchase value':group_total})
group_summary = group_summary.sort_values('purchase count', ascending = False)
group_summary["item price"] = group_summary["item price"].map('${:.2f}'.format)
group_summary['total purchase value'] = group_summary['total purchase value'].map('${:.2f}'.format) 
group_summary.head()
```
